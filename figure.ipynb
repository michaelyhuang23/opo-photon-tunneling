```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create the plot
fig, ax = plt.subplots()

# Plot the lines
line1, = ax.plot(x, y1, color='blue', linewidth=0.5, label='Simulated Trajectory')
line2, = ax.plot(x, y2, color='red', linewidth=3, label='Average Trajectory')

# Create the legend only
figlegend = plt.figure(figsize=(3, 2))  # New figure for legend
figlegend.legend(handles=[line1, line2], loc='center')  # Put legend in center

# Hide axes for the legend-only figure
plt.axis('off')

# Show the legend-only figure
plt.show()
```
