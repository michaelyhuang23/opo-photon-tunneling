```python
import matplotlib.pyplot as plt
from matplotlib import rc
rc('text', usetex=True)

plt.rcParams.update({
    'lines.linewidth': 2,       # Line thickness
    'lines.markersize': 10,     # Marker size
    'axes.linewidth': 1,        # Border thickness for axes
    'axes.titlesize': 20,       # Title font size
    'axes.labelsize': 20,       # Label font size
    'xtick.labelsize': 20,      # X-axis tick label font size
    'ytick.labelsize': 20,      # Y-axis tick label font size
    'xtick.major.size': 10,     # Length of X-axis ticks
    'xtick.major.width': 2,
    'xtick.minor.size': 5,     # Length of X-axis ticks
    'xtick.minor.width': 1,
    'ytick.major.size': 10,     # Length of X-axis ticks
    'ytick.major.width': 2,
    'ytick.minor.size': 5,     # Length of X-axis ticks
    'ytick.minor.width': 1,
    'legend.fontsize': 20,      # Legend font size
    'legend.frameon': True,     # Enable legend frame
    'legend.framealpha': 0.9,   # Legend frame transparency
    'legend.edgecolor': 'black', # Legend frame border color
})
```

```python
import json
with open('all_outputs.json') as f:
    data = json.load(f)
import pandas as pd
df = pd.DataFrame(data)
df.sort_values(by=['lambda', 'g', 'b'], inplace=True)
```

```python
plt.figure(figsize=(6, 3))
small_df = df[df['lambda']==2]
for g_val, color in zip([0.1, 0.15, 0.2, 0.25, 0.3, 0.4], ['orange', 'red', 'green', 'blue', 'purple', 'brown']):
    nano_df = small_df[small_df['g']==g_val]
    plt.plot(nano_df['b'], nano_df['sim_fit_time'], 'x', markerfacecolor=color, markeredgecolor=color, label=r"$g={}$".format(g_val))
    plt.plot(nano_df['b'], nano_df['sol1_time'], color=color)
    plt.yscale('log')
    plt.ylabel(r'\textit{time}')
    plt.xlabel(r'\textit{b}')
    plt.title(r"solution 1 $\lambda=2$")
    plt.legend()
plt.savefig('fix_lambda_sol1.png', format='png',  bbox_inches='tight')
```
